```python
   import pandas as pd

   # Load the data from CSV into a DataFrame
   data = pd.read_csv('fiveyearsweatherdata.csv')
```

```python
   # Check the dimensions of the DataFrame
   print("Dimensions of the data:", data.shape)

   # Preview the first few rows of the DataFrame
   print("First few rows of the data:")
   print(data.head())

   # Check for missing values
   print("Missing values in the data:")
   print(data.isnull().sum())
```

```python
   # Replace missing values with the mean of the respective columns
   data.fillna(data.mean(), inplace=True)
```

```python
   # Convert categorical variables into numerical
   data['Weather Type'] = data['Weather Type'].astype('category').cat.codes
```

```python
   # Calculate descriptive statistics of numerical columns
   print("Descriptive statistics:")
   print(data.describe())

   # Calculate correlation coefficients between features
   print("Correlation coefficients:")
   print(data.corr())
```

```python
   import matplotlib.pyplot as plt

   # Histogram of temperature
   plt.hist(data['Temperature'], bins=20)
   plt.xlabel('Temperature')
   plt.ylabel('Frequency')
   plt.title('Distribution of Temperature')
   plt.show()
```

```python
   import seaborn as sns

   # Scatter plot of temperature vs humidity
   sns.scatterplot(x='Temperature', y='Relative Humidity', data=data)
   plt.xlabel('Temperature')
   plt.ylabel('Relative Humidity')
   plt.title('Temperature vs Relative Humidity')
   plt.show()
```
